# Load Data Points (Do not modify the following block)

```python
with open('training_data.npz', 'rb') as f:
    data = np.load(f)

    x_list = data['x_list']
    y_list = data['y_list']

    x_data = data['x_data']
    y_data = data['y_data']
    n_data = len(x_data)
    
    w = data['w']
    original_degree = data['order']
    
# Print information of original function.
print("=================================")
print("We have", n_data, "number of data")

print("=================================")
weight_info_string = ''
for d in range(original_degree):
    weight_info_string += 'w'+str(d)+':'+str(round(w[d],ndigits=3))+'  '
print("Coefficients of the original polynomial")
print(weight_info_string)
print("=================================")
    
plt.plot(x_list, y_list, 'b:', linewidth=2, label="Original Function")
plt.scatter(x_data, y_data, s=50, c='r', label="Data Points")
plt.xlim([np.min(x_list),np.max(x_list)])
plt.ylim([np.min(y_data),np.max(y_data)])
plt.legend(prop={'size': 12})
plt.title("Data Plot")
plt.show()
```

# Polynomial Regression (Programming Assignment)
### Variable Explanation (Do not change variable names)
- 'w' is true coefficients of the original polynomial function
- 'original_degree' is the order of the original polynomial function
- 'x_list' is a list of the points at $x$-axis
- 'y_list' is a list of function value $f(x)$ corresponding to 'x_list'. In other words, y_list = $f($x_list$)$
- 'x_data' is an input data
- 'y_data' is an output data
- 'n_data' is the number of data points

### Our goal is to estimate 'w' from data points, 'x_data' and 'y_data'. Answer the following problems.

### 1. Compute a Vandermonde matrix when the degree of polynomial is $4$ (30pt)
- The variable 'degree' is the order of polynomial. In this problem, we set degree=$4$
- Use the variable 'A' for the Vandermonde matrix. Now, 'A' is initialized as a zero matrix whose elements are all zero. Fill in the element of the Vandermonde matrix by using power operator (\*\*), for loop, and np.concatenation.

```python
degree = 4
A = np.zeros((n_data, degree+1)) # Dummy initialization

for i in range(n_data):
    for j in range(degree+1):
        A[i, j] = x_data[i] ** j
```

### Print results (do not modify the following block)

```python
print(A)
```

### 2. Compute the coefficients of polynomial regression using a $4$ degree polynomial  (40pt)
- Use the variable 'degree' and the Vandermonde matrix 'A' in Problem 1.
- The variable 'w_est' is the coefficients of polynomial regression. Now, 'w_est' is initialized as a zero vector. Compute the 'w_est' from 'A' and 'y'
- The variable 'y_est' is an estimated function value corresponding to the input points 'x_list'. Now, it is a zero list and fill the list by computing the estimated function values. In other words, y_est = $\hat{f}($x_list$)$

```python
w_est = np.zeros((degree+1,1))
y_est = np.zeros_like(x_list)

# Compute w_est using the normal equation
ATA_inv = np.linalg.inv(np.dot(A.T, A))
ATy = np.dot(A.T, y_data)
w_est = np.dot(ATA_inv, ATy)

# Compute y_est for the x_list points
for i in range(len(x_list)):
    for j in range(degree+1):
        y_est[i] += w_est[j] * (x_list[i] ** j)
```

### Print results (do not modify the following block)

```python
plt.plot(x_list, y_list, 'b:', linewidth=2, label="Original Function")
plt.plot(x_list, y_est, 'm-', linewidth=2, label="Polynomial Regression (d={})".format(degree))
plt.scatter(x_data, y_data, s=50, c='r', label="Data Points")
plt.xlim([np.min(x_list),np.max(x_list)])
plt.ylim([np.min(y_data),np.max(y_data)])
plt.legend(prop={'size': 12})
plt.title("Data Plot")
plt.show()
```

## High Degree Polynomial:

### Pros:

Flexibility: A higher degree polynomial can model more complex and non-linear relationships.

Good Fit to Training Data: The model can closely fit the training data, potentially reducing training error.

### Cons:

Overfitting: High degree polynomials are susceptible to overfitting, which means they can capture noise in the training data, leading to poor generalization to new, unseen data.

Numerical Instability: As the degree increases, the coefficients can grow very large or very small, leading to potential numerical issues.

Interpretability: Higher degree polynomials can be harder to interpret than simpler models.

Increased Computation: Higher complexity can lead to more computational expense during both training and prediction.

## Low Degree Polynomial:

### Pros:

Simplicity: A lower degree polynomial is simpler and more interpretable.

Less Prone to Overfitting: With fewer parameters, the model is less likely to fit to noise in the data.

Stable: Numerically, it's more stable due to fewer coefficients.

### Cons:

Underfitting: A low degree polynomial might be too simple to capture the underlying relationship, especially if it's inherently non-linear. This can lead to high bias.

Poor Fit to Training Data: Might not capture all the patterns in the training data, leading to higher training error.

## Phenomenon in Machine Learning:

This balance between the model's complexity (like polynomial degree) and its performance on training vs. unseen data is referred to as the Bias-Variance Tradeoff in machine learning:

**Bias:** Error introduced by approximating a real-world problem (which might be complex) by a too-simple model. High bias can cause the model to miss relevant relations between features and target outputs, leading to underfitting.

**Variance:** Error introduced by the model's complexity. A model with high variance will perform very well on the training data but poorly on any unseen data, leading to overfitting.

In the context of polynomial regression, a high-degree polynomial might have low bias (as it fits the training data well) but high variance (as it reacts too much to fluctuations or noise in the training data). On the other hand, a low-degree polynomial might have high bias but low variance. Finding the right balance, often through techniques like cross-validation, is crucial to developing models that generalize well to new data.
